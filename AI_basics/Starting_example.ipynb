```python
#Example of classification with decision limit
#Author: U. Triltsch
#Last updated 3rd November 2024

import numpy as np
import matplotlib.pyplot as plt

learningrate = 1

################################
##Simple update function for decision boundary
# m: current slope
# b: current axis intercept
# x: list of x-points for line
# new_data_point: new data point
# label: label: 0: bolt; 1: nut
def update_m(m,b,x,new_datapoint,label):
    if label==0:  
        y_target=new_datapoint[1] - 1
    else:
        y_target=new_datapoint[1] + 1
    y_actual=new_datapoint[0] * m + b
    
    #fehler 
    error=(y_target - y_actual)
    
    delta_m = (error / new_datapoint[0]) * learningrate
    m_new = m + delta_m
    y_new = x * m_new + b
    plt.plot(x, y_new, 'c', linestyle='--')

# Bolts: percentage of black pixels, number of edges
bolts=np.array([(0.35, 30), (0.30, 25), (0.45, 32)])
print("Bolts array")
print(bolts)
print("----")


# Nuts: Proportion of black pixels, number of edges
nuts=np.array([(0.65, 25), (0.52, 21)])
print("Nuts array")
print(nuts)
print("----")
 
# Diagram formatting
plt.rcParams['lines.markersize'] = 8
plt.rcParams['lines.linewidth'] = 2
plt.axis([0, 1, 0, 40])
plt.xlabel('Percentage of black pixels')
plt.ylabel("Number of edges")
plt.grid()


# Show data points
for data_point in bolts:
    plt.plot(data_point[0], data_point[1],'bo')
for data_point in nuts:
    plt.plot(data_point[0], data_point[1],'rs')

    
'''  
# Show decision limit
x=np.linspace(0,1,10) # 10 equally spaced points between 0 and 1
print(x)

m=68.889
b=0
y_new = m * x + b

plt.plot(x, y_new, 'g', linestyle='--')
'''

'''
# new bolt
new_bolt = np.array([(0.4, 25)])
plt.plot(new_bolt[0][0], new_bolt[0][1], 'bo')
'''

'''
y_actual = m * new_bolt[0][0] + b
y_target = new_bolt[0][1] - 1


error = y_target - y_actual
delta_m = (error / new_bolt[0][0])
m_new = m + delta_m
y_new = x * m_new + b
plt.plot(x, y_new, 'c', linestyle='--')
'''

#update_m(m,b,x,new_bolt[0],0)

'''
#neue Mutter
new_bolt=np.array([(0.76,30)])
plt.plot(new_bolt[0][0], new_bolt[0][1], 'mo')

update_m(m, b, x, new_bolt[0], 1)
'''

plt.show()
```
